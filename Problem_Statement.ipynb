```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
monthly_sales_data_for_Product_A=[20, 30, 40, 50]
monthly_sales_data_for_Product_B=[25, 35, 45, 55]
```

```python

plt.plot(monthly_sales_data_for_Product_A)
plt.plot(monthly_sales_data_for_Product_B)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Data for Product A and Product B")
plt.legend(["Product A,Product B"])
plt.show()
```
